```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Column")
plt.ylabel("Missing Values")
plt.show()
```

```python
import numpy as np

books["missing_description"]= np.where(books["description"].isna(),1,0)
books['age_of_book'] = 2024-books["published_year" ]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"] #there are the to inspect the missing columns

correlation_matrix = books[columns_of_interest].corr(method = "spearman") # we used spearman instead of pearson because pearson deals with continuous data/values whereas spearman is more appropriate for discrete/non-continuous data

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm",
                      cbar_kws = {"label":"Spearman correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

```python
#this is the correlation heatmap
#anthing close to 1 or -1 has strong correlation and anything close to 0 has not weak or no relation
#for ex if we had strong correlation  with num pages and missing decription hence it would mean the longer or shorter books has more likely to have missing description
# so what we are going to do is just delete those entries that have missing description, rating
```

```python
#lets find out how many books we will need to delete
```

```python
book_missing  = books[~(books["description"].isna()) &
     ~(books["num_pages"].isna()) &
     ~(books["average_rating"].isna()) &
     ~(books["published_year"].isna())
]
```

```python
#we need enough information about the book so that we can recomend it
#And if it is not in the description then it won't be anywhere. So if some descrition is only one or 2 words won't be helpful  hence we will filter those out
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
#we are going to create a new variable the holds the number of  words in a given description of a book
book_missing
```

```python
#now we have added a new column "words in description
# "
book_missing.loc[book_missing["words_in_description"].between(15,24), "description"]
#what the .loc(row,col) method is it selects/locates the row containing 1 to 4 words in "words_in_Description" and the whole column in descripion and the whole thing will return ony the row that have words between 15 to 24 in the book description
```

```python
# we are going to create a filter code that filters out the rows that have less than 25 words in description

book_missing_25_words = book_missing[book_missing["words_in_description"] >=25]
book_missing_25_words
```

```python
#nOW Will be looking at subtitle
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
# as there a lot of subtitles that does not exts for boks. we are going to add a new column that has both the title and sub tite. If it has nly title it will show only ttle
```

```python
book_missing_25_words
```

```python
#create e new description for column for
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
# we create a new section that hold the isbn nad the description for toggether for our vector dataset use
```

```python
# drop all th column that we don't need
(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
